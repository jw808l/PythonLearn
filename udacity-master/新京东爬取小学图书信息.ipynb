```python
import urllib.request as urlrequest
from bs4 import BeautifulSoup as bs
import pandas as pd 
import numpy as np
```

```python
#获取小学类图书信息
#i为获取前几页的书籍
```

```python
def getBookInfo(page_len):
    lista=[]
    g=int(int(page_len)+1)
    for page in range(1,g):
        u1='https://search.jd.com/Search?keyword=%E5%B0%8F%E5%AD%A6&enc=utf-8&qrst=1&rt=1&stop=1&vt=2&wq=%E5%B0%8F%E5%AD%A6&psort=3&page='
        u2=str(int(page*2-1))
        url=u1+u2
        #获取页面信息
        web_page=urlrequest.urlopen(url).read()
        web_soup=bs(web_page,'html.parser')
        
        #获取相应标签的值
        names = web_soup.find_all(class_="p-name")
        origins=web_soup.find_all(class_='curr-shop hd-shopname')
        prices=web_soup.find_all(class_='p-price')
        urls=web_soup.find_all(class_='p-img')
        
        #将每夜的信息记录在bookinfo的字典中

        for i,j,k,l in zip(names,origins,prices,urls):
            bookinfo={}
            bookinfo['name']=i.find('em').text.strip()
            bookinfo['origin']=j.get_text()
            bookinfo['price']=k.get_text()
            bookinfo['url']=l.a['href']
            lista.append(bookinfo)
    return lista
```

```python
#获取前四页的书籍信息
bookinfo=pd.DataFrame(getBookInfo(4))
bookinfo
```
